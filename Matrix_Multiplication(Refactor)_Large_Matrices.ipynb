```python
import duckdb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import psutil
import os
from matplotlib.ticker import FuncFormatter

# Set up for plots
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = [14, 8]
plt.rcParams['font.size'] = 12

def get_memory_usage():
    """Get current memory usage in MB"""
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / (1024 * 1024)  # Convert to MB

def benchmark_matrix_multiplication(A, B, method_name="DuckDB inner_product"):
    """Benchmark matrix multiplication using different methods"""

    print(f" MATRIX MULTIPLICATION BENCHMARK ({A.shape[0]}x{A.shape[1]} * {B.shape[0]}x{B.shape[1]})")
    print("=" * 80)

    # Benchmark results storage
    benchmarks = {
        'DuckDB inner_product': {'time': 0, 'memory': 0, 'result': None},
        'NumPy einsum': {'time': 0, 'memory': 0, 'result': None},
        'NumPy matmul': {'time': 0, 'memory': 0, 'result': None}
    }

    # Benchmark 1: DuckDB inner_product
    print("\n1.  DUCKDB INNER_PRODUCT")
    print("-" * 40)

    # Get matrix dimensions
    m, k = A.shape
    k, p = B.shape

    # DuckDB matrix multiplication using array_inner_product
    start_mem = get_memory_usage()
    start_time = time.perf_counter()

    try:
        # Connect and register with DuckDB
        con = duckdb.connect()

        # Create DuckDB tables with explicit fixed-size ARRAY types
        con.execute(f"CREATE TABLE a (row_id INTEGER, row DOUBLE[{k}])")
        for i, row_data in enumerate(A.tolist()):
            con.execute(f"INSERT INTO a VALUES (?, ?)", (i, row_data))

        # For B_T, each row is also a DOUBLE[k] (fixed size)
        B_T = B.T
        con.execute(f"CREATE TABLE bt (col_id INTEGER, row DOUBLE[{k}])")
        for i, col_data in enumerate(B_T.tolist()):
            con.execute(f"INSERT INTO bt VALUES (?, ?)", (i, col_data))

        # Execute the query
        result_df = con.execute("""
        SELECT
          a.row_id AS row,
          bt.col_id AS col,
          array_inner_product(a.row, bt.row) AS value
        FROM a
        CROSS JOIN bt
        ORDER BY row, col
        """).fetchdf()

        end_time = time.perf_counter()
        end_mem = get_memory_usage()

        benchmarks['DuckDB inner_product']['time'] = end_time - start_time
        benchmarks['DuckDB inner_product']['memory'] = end_mem - start_mem

        # Convert result to matrix
        rows = result_df['row'].max() + 1
        cols = result_df['col'].max() + 1
        duckdb_result = np.zeros((rows, cols))
        for _, r in result_df.iterrows():
            duckdb_result[int(r['row']), int(r['col'])] = r['value']

        benchmarks['DuckDB inner_product']['result'] = duckdb_result

        print(f"   Result shape: {duckdb_result.shape}")
        print(f"   Time: {benchmarks['DuckDB inner_product']['time']:.6f} seconds")
        print(f"   Memory: {benchmarks['DuckDB inner_product']['memory']:.2f} MB")

    except Exception as e:
        print(f"   DuckDB failed with error: {e}")
        benchmarks['DuckDB inner_product']['time'] = float('inf')
        benchmarks['DuckDB inner_product']['memory'] = float('inf')

    finally:
        if 'con' in locals():
            con.close()

    # Benchmark 2: NumPy einsum
    print("\n2.  NUMPY EINSUM")
    print("-" * 40)

    try:
        start_mem = get_memory_usage()
        start_time = time.perf_counter()
        numpy_einsum_result = np.einsum('ik,kj->ij', A, B)
        end_time = time.perf_counter()
        end_mem = get_memory_usage()

        benchmarks['NumPy einsum']['time'] = end_time - start_time
        benchmarks['NumPy einsum']['memory'] = end_mem - start_mem
        benchmarks['NumPy einsum']['result'] = numpy_einsum_result

        print(f"   Result shape: {numpy_einsum_result.shape}")
        print(f"   Time: {benchmarks['NumPy einsum']['time']:.6f} seconds")
        print(f"   Memory: {benchmarks['NumPy einsum']['memory']:.2f} MB")

    except Exception as e:
        print(f"   NumPy einsum failed with error: {e}")
        benchmarks['NumPy einsum']['time'] = float('inf')
        benchmarks['NumPy einsum']['memory'] = float('inf')

    # Benchmark 3: NumPy matmul
    print("\n3.  NUMPY MATMUL")
    print("-" * 40)

    try:
        start_mem = get_memory_usage()
        start_time = time.perf_counter()
        numpy_matmul_result = np.matmul(A, B)
        end_time = time.perf_counter()
        end_mem = get_memory_usage()

        benchmarks['NumPy matmul']['time'] = end_time - start_time
        benchmarks['NumPy matmul']['memory'] = end_mem - start_mem
        benchmarks['NumPy matmul']['result'] = numpy_matmul_result

        print(f"   Result shape: {numpy_matmul_result.shape}")
        print(f"   Time: {benchmarks['NumPy matmul']['time']:.6f} seconds")
        print(f"   Memory: {benchmarks['NumPy matmul']['memory']:.2f} MB")

    except Exception as e:
        print(f"   NumPy matmul failed with error: {e}")
        benchmarks['NumPy matmul']['time'] = float('inf')
        benchmarks['NumPy matmul']['memory'] = float('inf')

    # Verify results match if all succeeded
    print("\n4. âœ… VERIFICATION")
    print("-" * 40)

    if (benchmarks['DuckDB inner_product']['result'] is not None and
        benchmarks['NumPy einsum']['result'] is not None and
        benchmarks['NumPy matmul']['result'] is not None):

        match_einsum = np.allclose(duckdb_result, numpy_einsum_result)
        match_matmul = np.allclose(duckdb_result, numpy_matmul_result)

        print(f"   DuckDB vs NumPy einsum: {match_einsum}")
        print(f"   DuckDB vs NumPy matmul: {match_matmul}")
    else:
        print("   Verification skipped due to failures in some methods")

    # Performance comparison
    print("\n5.  PERFORMANCE COMPARISON")
    print("-" * 40)

    if (benchmarks['DuckDB inner_product']['time'] != float('inf') and
        benchmarks['NumPy einsum']['time'] != float('inf')):
        speed_ratio_einsum = benchmarks['NumPy einsum']['time'] / benchmarks['DuckDB inner_product']['time']
        print(f"    DuckDB vs NumPy einsum: {speed_ratio_einsum:.2f}x {'faster' if speed_ratio_einsum > 1 else 'slower'}")
    else:
        print("    DuckDB vs NumPy einsum: Comparison not available")

    if (benchmarks['DuckDB inner_product']['time'] != float('inf') and
        benchmarks['NumPy matmul']['time'] != float('inf')):
        speed_ratio_matmul = benchmarks['NumPy matmul']['time'] / benchmarks['DuckDB inner_product']['time']
        print(f"    DuckDB vs NumPy matmul: {speed_ratio_matmul:.2f}x {'faster' if speed_ratio_matmul > 1 else 'slower'}")
    else:
        print("    DuckDB vs NumPy matmul: Comparison not available")

    return benchmarks

def create_individual_comparison(benchmarks, size_label, all_benchmarks_history):
    """Create line graph comparison for a single matrix size with historical context"""

    methods = ['DuckDB inner_product', 'NumPy einsum', 'NumPy matmul']
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
    markers = ['o', 's', '^']

    # Extract historical data
    matrix_sizes = []
    time_data = {method: [] for method in methods}
    memory_data = {method: [] for method in methods}

    for label, bench_data in all_benchmarks_history.items():
        # Extract matrix dimensions from label
        size_str = label.split('(')[1].split(')')[0]
        matrix_sizes.append(size_str)

        for method in methods:
            time_data[method].append(bench_data[method]['time'])
            memory_data[method].append(bench_data[method]['memory'])

    # Create subplots with two y-axes for memory to handle different scales
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Time comparison
    for i, method in enumerate(methods):
        # Filter out infinite values
        valid_times = [t if t != float('inf') else None for t in time_data[method]]
        ax1.plot(range(len(matrix_sizes)), valid_times,
                label=method, color=colors[i], marker=markers[i], linewidth=2, markersize=8)

    ax1.set_xlabel('Matrix Size')
    ax1.set_ylabel('Execution Time (seconds)')
    ax1.set_title(f'Execution Time Progression\nCurrent: {size_label}', fontsize=14, fontweight='bold')
    ax1.set_xticks(range(len(matrix_sizes)))
    ax1.set_xticklabels(matrix_sizes, rotation=45, ha='right')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Set y-axis to log scale for better visualization of wide range
    ax1.set_yscale('log')

    # Memory comparison with dual y-axis
    ax2_mem = ax2.twinx()  # Create a second y-axis for memory

    # Plot DuckDB memory on left y-axis
    duckdb_mem = [m if m != float('inf') else None for m in memory_data['DuckDB inner_product']]
    line1 = ax2.plot(range(len(matrix_sizes)), duckdb_mem,
                    label='DuckDB Memory', color='#FF6B6B', marker='o', linewidth=2, markersize=8)

    # Plot NumPy memory on right y-axis
    numpy_einsum_mem = [m if m != float('inf') else None for m in memory_data['NumPy einsum']]
    numpy_matmul_mem = [m if m != float('inf') else None for m in memory_data['NumPy matmul']]

    # Add a small epsilon to avoid zero values for log scale
    epsilon = 1e-10
    numpy_einsum_mem = [max(m, epsilon) if m is not None else None for m in numpy_einsum_mem]
    numpy_matmul_mem = [max(m, epsilon) if m is not None else None for m in numpy_matmul_mem]

    line2 = ax2_mem.plot(range(len(matrix_sizes)), numpy_einsum_mem,
                        label='NumPy einsum Memory', color='#4ECDC4', marker='s', linewidth=2, markersize=8)
    line3 = ax2_mem.plot(range(len(matrix_sizes)), numpy_matmul_mem,
                        label='NumPy matmul Memory', color='#45B7D1', marker='^', linewidth=2, markersize=8)

    ax2.set_xlabel('Matrix Size')
    ax2.set_ylabel('DuckDB Memory Usage (MB)', color='#FF6B6B')
    ax2_mem.set_ylabel('NumPy Memory Usage (MB)', color='#45B7D1')
    ax2.set_title(f'Memory Usage Progression\nCurrent: {size_label}', fontsize=14, fontweight='bold')
    ax2.set_xticks(range(len(matrix_sizes)))
    ax2.set_xticklabels(matrix_sizes, rotation=45, ha='right')

    # Set y-axes to log scale for better visualization, but only if we have positive values
    if any(m > 0 for m in duckdb_mem if m is not None):
        ax2.set_yscale('log')

    if any(m > 0 for m in numpy_einsum_mem + numpy_matmul_mem if m is not None):
        ax2_mem.set_yscale('log')

    # Combine legends from both axes
    lines, labels = ax2.get_legend_handles_labels()
    lines2, labels2 = ax2_mem.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='upper left')

    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

def create_final_progression_comparison(all_benchmarks):
    """Create final progression visualization showing performance across all matrix sizes"""

    methods = ['DuckDB inner_product', 'NumPy einsum', 'NumPy matmul']
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
    markers = ['o', 's', '^']

    # Extract matrix sizes from labels
    matrix_sizes = []
    time_data = {method: [] for method in methods}
    memory_data = {method: [] for method in methods}

    for label, benchmarks in all_benchmarks.items():
        # Extract matrix dimensions from label
        size_str = label.split('(')[1].split(')')[0]
        matrix_sizes.append(size_str)

        for method in methods:
            time_data[method].append(benchmarks[method]['time'])
            memory_data[method].append(benchmarks[method]['memory'])

    # Create subplots with dual y-axis for memory
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Time comparison
    for i, method in enumerate(methods):
        # Filter out infinite values
        valid_times = [t if t != float('inf') else None for t in time_data[method]]
        ax1.plot(range(len(matrix_sizes)), valid_times,
                label=method, color=colors[i], marker=markers[i], linewidth=2, markersize=8)

    ax1.set_xlabel('Matrix Size')
    ax1.set_ylabel('Execution Time (seconds)')
    ax1.set_title('Final Execution Time Progression', fontsize=14, fontweight='bold')
    ax1.set_xticks(range(len(matrix_sizes)))
    ax1.set_xticklabels(matrix_sizes, rotation=45, ha='right')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Set y-axis to log scale for better visualization
    ax1.set_yscale('log')

    # Memory comparison with dual y-axis
    ax2_mem = ax2.twinx()  # Create a second y-axis for memory

    # Plot DuckDB memory on left y-axis
    duckdb_mem = [m if m != float('inf') else None for m in memory_data['DuckDB inner_product']]
    line1 = ax2.plot(range(len(matrix_sizes)), duckdb_mem,
                    label='DuckDB Memory', color='#FF6B6B', marker='o', linewidth=2, markersize=8)

    # Plot NumPy memory on right y-axis
    numpy_einsum_mem = [m if m != float('inf') else None for m in memory_data['NumPy einsum']]
    numpy_matmul_mem = [m if m != float('inf') else None for m in memory_data['NumPy matmul']]

    # Add a small epsilon to avoid zero values for log scale
    epsilon = 1e-10
    numpy_einsum_mem = [max(m, epsilon) if m is not None else None for m in numpy_einsum_mem]
    numpy_matmul_mem = [max(m, epsilon) if m is not None else None for m in numpy_matmul_mem]

    line2 = ax2_mem.plot(range(len(matrix_sizes)), numpy_einsum_mem,
                        label='NumPy einsum Memory', color='#4ECDC4', marker='s', linewidth=2, markersize=8)
    line3 = ax2_mem.plot(range(len(matrix_sizes)), numpy_matmul_mem,
                        label='NumPy matmul Memory', color='#45B7D1', marker='^', linewidth=2, markersize=8)

    ax2.set_xlabel('Matrix Size')
    ax2.set_ylabel('DuckDB Memory Usage (MB)', color='#FF6B6B')
    ax2_mem.set_ylabel('NumPy Memory Usage (MB)', color='#45B7D1')
    ax2.set_title('Final Memory Usage Progression', fontsize=14, fontweight='bold')
    ax2.set_xticks(range(len(matrix_sizes)))
    ax2.set_xticklabels(matrix_sizes, rotation=45, ha='right')

    # Set y-axes to log scale for better visualization, but only if we have positive values
    if any(m > 0 for m in duckdb_mem if m is not None):
        ax2.set_yscale('log')

    if any(m > 0 for m in numpy_einsum_mem + numpy_matmul_mem if m is not None):
        ax2_mem.set_yscale('log')

    # Combine legends from both axes
    lines, labels = ax2.get_legend_handles_labels()
    lines2, labels2 = ax2_mem.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='upper left')

    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('matrix_multiplication_progression.png', dpi=300, bbox_inches='tight')
    plt.show()

def run_matrix_multiplication_benchmark():
    """Run matrix multiplication benchmark with different matrix sizes"""

    # Test with different matrix sizes - significantly increased sizes
    matrix_sizes = [
        (2, 3, 3, 2),           # Very small matrices
        (5, 10, 10, 5),         # Small matrices
        (10, 20, 20, 10),       # Small-medium matrices
        (20, 30, 30, 20),       # Medium matrices
        (50, 100, 100, 50),     # Medium-large matrices
        (100, 200, 200, 100),   # Large matrices
        (200, 300, 300, 200),   # Larger matrices
        (300, 400, 400, 300),   # Very large matrices
        (500, 600, 600, 500),   # Extremely large matrices
        (800, 1000, 1000, 800), # Very large matrices 2
        (1000, 1200, 1200, 1000), # Massive matrices
        (1500, 1800, 1800, 1500) # Very massive matrices
    ]

    size_labels = [
        "Very Small Matrices (2x3 * 3x2)",
        "Small Matrices (5x10 * 10x5)",
        "Small-Medium Matrices (10x20 * 20x10)",
        "Medium Matrices (20x30 * 30x20)",
        "Medium-Large Matrices (50x100 * 100x50)",
        "Large Matrices (100x200 * 200x100)",
        "Larger Matrices (200x300 * 300x200)",
        "Very Large Matrices (300x400 * 400x300)",
        "Extremely Large Matrices (500x600 * 600x500)",
        "Very Large Matrices 2 (800x1000 * 1000x800)",
        "Massive Matrices (1000x1200 * 1200x1000)",
        "Very Massive Matrices (1500x1800 * 1800x1500)"
    ]

    all_benchmarks = {}

    for i, ((m, k, n, p), label) in enumerate(zip(matrix_sizes, size_labels)):
        print(f"\n{'='*100}")
        print(f"BENCHMARKING MATRIX MULTIPLICATION {i+1}: {label}")
        print(f"{'='*100}")

        # Check if we should skip this size due to memory constraints
        # Estimate memory usage: A (m*k*8 bytes) + B (n*p*8 bytes) + result (m*p*8 bytes)
        estimated_memory = (m*k + n*p + m*p) * 8 / (1024**2)  # Convert to MB

        print(f"Estimated memory usage: {estimated_memory:.2f} MB")

        if estimated_memory > 4000:  # Skip if estimated memory > 4GB
            print("Skipping this size due to high memory requirements")
            benchmarks = {
                'DuckDB inner_product': {'time': float('inf'), 'memory': float('inf'), 'result': None},
                'NumPy einsum': {'time': float('inf'), 'memory': float('inf'), 'result': None},
                'NumPy matmul': {'time': float('inf'), 'memory': float('inf'), 'result': None}
            }
        else:
            # Generate random matrices
            np.random.seed(42)
            A = np.random.rand(m, k)
            B = np.random.rand(n, p)

            # Run benchmark
            benchmarks = benchmark_matrix_multiplication(A, B)

        all_benchmarks[label] = benchmarks

        # Create line graph comparison with historical data
        create_individual_comparison(benchmarks, label, all_benchmarks)

        # Print detailed results for this matrix size
        print("\n" + "="*120)
        print(f"DETAILED RESULTS: {label}")
        print("="*120)
        methods = ['DuckDB inner_product', 'NumPy einsum', 'NumPy matmul']
        print(f"{'Method':<25} {'Time (s)':<12} {'Memory (MB)':<12} {'Speed Ratio':<12}")
        print("-"*120)

        for method in methods:
            time_val = benchmarks[method]['time']
            memory_val = benchmarks[method]['memory']

            if time_val == float('inf') or memory_val == float('inf'):
                time_str = "FAILED"
                memory_str = "FAILED"
                ratio_str = "N/A"
            else:
                time_str = f"{time_val:<12.6f}"
                memory_str = f"{memory_val:<12.2f}"
                if method == 'DuckDB inner_product':
                    ratio = 1.0
                else:
                    ratio = time_val / benchmarks['DuckDB inner_product']['time'] if benchmarks['DuckDB inner_product']['time'] != float('inf') else float('inf')
                ratio_str = f"{ratio:<12.2f}" if ratio != float('inf') else "N/A"

            print(f"{method:<25} {time_str} {memory_str} {ratio_str}")

    # Create final progression comparison after all benchmarks
    print("\n" + "="*140)
    print("FINAL PROGRESSION COMPARISON ACROSS ALL MATRIX SIZES")
    print("="*140)
    create_final_progression_comparison(all_benchmarks)

    return all_benchmarks

# Run the benchmark
if __name__ == "__main__":
    results = run_matrix_multiplication_benchmark()
```
